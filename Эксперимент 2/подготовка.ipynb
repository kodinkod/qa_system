## SPLIT CONTENT

#### Разобьем данне в исходных  файлах - Просто добаим новые Headline в документации

```python
import murgpt
import pandas as pd
```

```python
PATCH = "../Примеры инструкций/ДозорPREP/"
EXTENSIONS = ['.docx']
files = murgpt.get_list_files(PATCH , ['.docx'])
INFO  = murgpt.read_docs(files)

predata = {'title':[], 'content':[]}
for k in INFO['content']:
    if INFO["content"][k] != '':
        predata['title'].append(k)
        predata['content'].append(k + ' \n ' + INFO['content'][k].replace("\n", "\n " ))

data = pd.DataFrame(predata)
```

```python
data.loc[0].title = "Общая информация о 'Дозор'"
data
```

```python
data['count'] = [murgpt.get_counts_token(encoding, item) for item in data['content']]
```

## Дополняем TL;DR

```python
def make_tl_dr(text):
    mes = [{"role": "system", "content": "Ты помошник по обработке текста!"},
           {"role": "user", "content": "Напиши краткий пересказ текста"},
           {"role": "user", "content": text + "TL;DR"}]

    pre_text = murgpt.generate_answer_from_promt(mes)
    return pre_text
    
```
