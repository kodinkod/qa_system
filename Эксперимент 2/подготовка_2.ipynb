### Добавим для поиска <b>threshold</b> : <font style="color:blue;">create_context_3ver</font>

```python
import murgpt
import pandas as pd
data = pd.read_csv('data_dozor/DATAMAIN.csv')
data = data.drop('Unnamed: 0', axis=1)
data['embeddings'] = data.embeddings.apply(murgpt.emb_to_list) #делает эмбеддинги нормальными в пандас
data['embeddings_tl_dr'] = data.embeddings_tl_dr.apply(murgpt.emb_to_list) #делает эмбеддинги нормальными в пандас
data['embeddings_qa'] = data.embeddings_qa.apply(murgpt.emb_to_list) #делает эмбеддинги нормальными в пандас
```

```python
question = "Где можно купить булочку хлеба?"       # min = 0.24
#question = "как установить Дозор?"                 # min = 0.148
#question = "Можно ли установить на виндовс"        # min = 0.204
#question = "Как активировать"                      # min = 0.16
#question = "Мне подойдет процессор intel core i3"   # min = 0.201
```

```python
murgpt.create_context_3ver(question, 
                           data, max_len=1800, 
                           cont_name = 'content', 
                           emb_name = 'embeddings_tl_dr', 
                           count_name = 'count', 
                                           # максимальное расстояние при котором подсказка отсекается 
                           )
```
