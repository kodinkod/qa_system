```python
INFO  = murgpt.read_docs(files)
INFO['content']
```

```python
INFO['struct']
```

```python
predata = {'title':[], 'content':[]}
for k in INFO['content']:
    if INFO["content"][k] != '':
        predata['title'].append(k)
        predata['content'].append(k + '\n ' + INFO['content'][k])

data = pd.DataFrame(predata)
```
